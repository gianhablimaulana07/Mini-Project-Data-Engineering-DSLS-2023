# Employee Analysis

## Analisis hubungan employee dengan banyaknya order.

Analisis hubungan employee dengan banyaknya order dilakukan untuk menentukan pegawai terbaik yang memberikan sumbangsih penjualan terbesar dan menentukan jabatan pegawai mana yang paling berkontribusi dalam penjualan.

## Import Library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Import Dataset

Dataset diperoleh dari hasil querry pada database Northwind sebagai berikut.

!

```python
df = pd.read_csv("Employee Order Table.csv",encoding="latin-1", sep=';')
df
```

## Visualisasi

```python
plt.figure(figsize=[12,8])

sns.barplot(x='Employee_Name', y='Number_of_Order', data=df, palette='ocean')
plt.title('Number of Orders by Employee Name')
plt.xlabel('Employee_Name')
plt.xticks(rotation=45)
plt.ylabel('Number_of_Order')
plt.show()
```

```python
df_agg = df.groupby('Title').agg({'Number_of_Order': 'sum'})
df_agg.columns = ['Total_Number_of_Order']

df_agg
```

```python
plt.figure(figsize=[12,8])

sns.barplot(x=df_agg.index, y='Total_Number_of_Order', data=df_agg, palette='spring')
plt.title('Number of Orders by Title')
plt.xlabel('Title')
plt.xticks(rotation=45)
plt.ylabel('Number_of_Order')
plt.show()
```

```python
plt.figure(figsize=[12,8])

sns.barplot(x='Employee_Name', y='Number_of_Discount', data=df, palette='summer')
plt.title('Number_of_Discount by Employee Name')
plt.xlabel('Employee_Name')
plt.xticks(rotation=45)
plt.ylabel('Number_of_Discount')
plt.show()
```

```python
df_agg1 = df.groupby('Title').agg({'Number_of_Discount': 'sum'})
df_agg1.columns = ['Total_Number_of_Discount']

df_agg1
```

```python
plt.figure(figsize=[12,8])

sns.barplot(x=df_agg.index, y='Total_Number_of_Discount', data=df_agg1, palette='tab20')
plt.title('Total_Number_of_Discount by Title')
plt.xlabel('Title')
plt.xticks(rotation=45)
plt.ylabel('Total_Number_of_Discount')
plt.show()
```
